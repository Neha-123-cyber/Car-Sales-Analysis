```python
## Plotting a bar chart for gender it's count

ax = sns.countplot(x = 'Gender',data = s)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
From above graphs we can see that most of the buyers are  Male and even the purchasing power of Male are greater than Female
```

```python
sales_Engine = s.groupby(['Engine'], as_index=False)['Annual Income'].sum().sort_values(by='Annual Income', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_Engine, x = 'Engine',y= 'Annual Income')
```

```python
sns.countplot(x = 'Date' , hue = 'Transmission', data = s)
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))
s.groupby('Car_id')['Price ($)'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```

```python
Dealer_Region = s.groupby(['Dealer_Region'], as_index=False)['Price ($)'].sum().sort_values(by='Price ($)', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = Dealer_Region, x = 'Dealer_Region',y= 'Price ($)')
```
